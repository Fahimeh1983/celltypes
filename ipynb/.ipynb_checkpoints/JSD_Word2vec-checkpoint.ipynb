Here we take only the upper triangle of the JSD matrix. The new data frame which has the type1---type2---weights can be used for creating the graph.

```python
upper_tri = np.triu(np.ones(sim.shape), 0).astype(np.bool)
type_edge_weight = sim.where(upper_tri)
type_edge_weight = type_edge_weight.stack().reset_index()
type_edge_weight.columns = ['type1','type2','weight']
type_edge_weight
```

### Find the most similar node to a given node

```python
given_type = 'Sst Rxfp1 Prdm8'
dist_from_node = []

for i in types:
    dist_from_node.append((i, cosine_distance[(given_type, i)]))

def Sort_Tuple(tup):  
    lst = len(tup)  
    for i in range(0, lst):  
        for j in range(0, lst-i-1):  
            if (tup[j][1] > tup[j + 1][1]):  
                temp = tup[j]  
                tup[j]= tup[j + 1]  
                tup[j + 1]= temp  
    return tup  

Sort_Tuple(dist_from_node)[1:10]
```
